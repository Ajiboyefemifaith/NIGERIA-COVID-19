```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
%matplotlib inline
```

```python
os.getcwd()
```

```python
data=pd.read_csv('Nigeria_covid19.csv')
```

```python
data.head()
```

```python
data.columns=['States', 'TotalCases', 'ActiveCases', 'Recovered', 'TotalDeaths']
```

```python
data.info()
```

VISUALIZATION FOR TOTAL COVID19 CASES IN NIGERIA

```python
sns.barplot(data=data, x='TotalCases', y='States')
sns.set_style('darkgrid')
plt.rcParams['figure.figsize']=15,8
plt.title('COVID-19 CASES IN NIGERIA', fontsize=20)
plt.show()
```

VISUALIZATION FOR ACTIVE COVID19 CASES IN NIGERIA

```python
sns.barplot(data=data, x='ActiveCases', y='States')
plt.title('ACTIVE COVID-19 CASES IN NIGERIA', fontsize=20)
plt.show()
```

VISUALIZATION FOR RECOVERED COVID19 CASES IN NIGERIA

```python
sns.barplot(data=data, x='Recovered', y='States')
plt.title('RECOVERED COVID-19 CASES IN NIGERIA', fontsize=20)
plt.show()
```

VISUALIZATION FOR DEATH COVID19 CASES IN NIGERIA

```python
sns.barplot(data=data, x='TotalDeaths', y='States')
plt.title('DEATH COVID-19 CASES IN NIGERIA', fontsize=20)
plt.show()
```

RECOVERY RATE

```python
#To add Recovery_rate and Death_rate columns to the dataframe and round its values to 2 decimal places.
data['Recovery_rate']=round(data['Recovered']/data['TotalCases']*100,2)
data['Death_rate']=round(data['TotalDeaths']/data['TotalCases']*100,2)
```

```python
data.head()
```

RECOVERY RATE FOR ALL STATES

```python
sns.barplot(data=data, y='States',x= 'Recovery_rate')
plt.title('RECOVERY RATE FOR COVID-19', fontsize=20)
plt.show()
```

DEATH RATE FOR ALL STATES

```python
sns.barplot(data=data, y='States',x= 'Death_rate')
plt.title('DEATH RATE FOR COVID-19', fontsize=20)
plt.show()
```

TOP 10 CASES

```python
m1=data[['States', 'TotalCases']].sort_values('TotalCases',ascending=False).nlargest(10,'TotalCases')
```

```python
sns.barplot(data=m1, y='States',x= 'TotalCases')
plt.title('TOP 10 STATES WITH COVID-19', fontsize=20)
plt.show()
```

```python
sns.barplot(data=m3, y='States',x= 'Recovered')
plt.title('TOP 10 STATES WITH RECOVERED COVID-19 CASES', fontsize=20)
plt.show()
```

```python
m4=data[['States', 'ActiveCases']].sort_values('ActiveCases',ascending=False).nlargest(10,'ActiveCases')
```

```python
sns.barplot(data=m4, y='States',x= 'ActiveCases')
plt.title('TOP 10 STATES WITH ACTIVE COVID-19 CASES', fontsize=20)
plt.show()
```

```python
m5=data[['States', 'Recovery_rate']].sort_values('Recovery_rate',ascending=False).nlargest(10,'Recovery_rate')
```

```python
sns.barplot(data=m5, y='States',x= 'Recovery_rate')
plt.title('RECOVERY RATE FOR TOP 10 STATES WITH COVID-19 CASES', fontsize=20)
plt.show()
```

```python
sns.barplot(data=m5, y='States',x= 'Death_rate')
plt.title('DEATH RATE FOR TOP 10 STATES WITH COVID-19 CASES', fontsize=20)
plt.show()
```

DEATH AND RECOVERY RATE DISTRIBUTION ACROSS TOP TEN STATES IN NIGERIA

```python
plt.hist(data=data, x='Recovery_rate',bins=15)
plt.ylabel('COUNT',fontsize=15)
plt.xlabel('Recovery_rate',fontsize=15)
plt.title('COVID-19RECOVERY RATE DISTRRIBUTION ACROSS TOP TEN STATES IN NIGERIA',fontsize=15)
plt.show()
```

```python
plt.hist(data=data, x='Death_rate',bins=15)
plt.ylabel('COUNT',fontsize=15)
plt.xlabel('Death_rate',fontsize=15)
plt.title('COVID-19 DEATH RATE DISTRRIBUTION ACROSS TOP TEN STATES IN NIGERIA',fontsize=15)
plt.show()
```
